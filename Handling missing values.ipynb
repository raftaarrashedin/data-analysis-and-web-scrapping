MCAR stands for Missing Completely At Random, which is one of the mechanisms that describe the missingness of data in statistical analysis and data handling.

When data is missing completely at random, it means that the likelihood of a data point being missing is unrelated to both observed and unobserved data. In other words, there is no systematic pattern or reason for why certain data points are missing.

Key characteristics of MCAR:

    Randomness: The missing data points are missing due to random chance, and there is no pattern or bias in their absence.

    No Dependency: There is no relationship between the missing data and any other variables in the dataset.

    Missing Data Not Predictable: You cannot predict whether a particular data point will be missing based on the values of other variables in the dataset.

Dealing with MCAR data is often less problematic than dealing with data missing under other mechanisms (e.g., Missing Not At Random or MNAR). In the case of MCAR, common techniques for handling missing data include listwise deletion (removing rows with missing data), mean imputation, or more advanced methods like multiple imputation, depending on the specific analysis and data context.

It's important to identify the mechanism of missing data because the method chosen to handle missing data may vary depending on whether the data is MCAR, MNAR, or Missing At Random (MAR), where missingness depends on observed data but not unobserved data. Handling missing data appropriately is crucial to avoid bias and ensure the validity of statistical analyses.

Example: Survey Data

Suppose you are conducting a survey about people's favorite ice cream flavors. You collect responses from 100 participants, and the survey includes questions about age, gender, and their preferred ice cream flavor. However, due to various reasons such as data entry errors or survey non-response, you end up with some missing data for the age variable.

Here's how MCAR applies in this scenario:

    Randomness: The missing age data points occur randomly, without any particular pattern. For example, some participants might have skipped the age question, but there's no systematic reason like "older people are more likely to skip the question."

    No Dependency: There's no relationship between the missing age data and any other variables in the dataset. The fact that a participant's age is missing has no connection to their gender or ice cream preference.

    Missing Data Not Predictable: You can't predict which participants will have missing age data based on their gender, ice cream preference, or any other information in the dataset.

In this case, if the missing data is MCAR, you can use various methods to handle it, such as listwise deletion (removing rows with missing age data), mean imputation (replacing missing ages with the mean age of the respondents), or more sophisticated techniques if appropriate for your analysis.

**RANDOM SAMPLE IMPUTATION**

Random Sample Imputation consists of taking random observsation from the dataset and we use this observation to replace NaN values.

When shoud It be used> It assumes that the data are missing completely at random(MCAR).

```python
df = pd.read_csv('/content/tested.csv', usecols = ['Age','Fare','Survived'])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.isnull().mean()
```

```python
df['Age'].isnull().sum()
```

```python
df['Age'].dropna().sample(df['Age'].isnull().sum(), random_state = 0)
```

```python
df[df['Age'].isnull()].index
```

```python
def impute_Nan(df,variable,median) :

  # creating two features
  df[variable+"_median"] = df[variable].fillna(median)
  df[variable+"_random"] = df[variable]

  # random sample to fill Na
  # picking some random sample
  random_sample = df[variable].dropna().sample(df[variable].isnull().sum(), random_state = 42)
  # pandas need to have same index in order to merge the dataset
  # finding indexes of null values to replace NaN value
  random_sample.index = df[df[variable].isnull()].index
  # replacing null or Nan values
  df.loc[df[variable].isnull(),variable+'_random'] = random_sample
```

```python
median = df.Age.median
```

```python
median
```

```python
impute_Nan(df,'Age',median)
```

```python
df.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
df['Age'].plot(kind = 'kde', ax = ax, color = 'green')
df.Age_random.plot(kind = 'kde', ax = ax, color = 'black')
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc = 'best')
```

1. Easy to implement
2. less distortion



Disadvantages


1. In every situation the randomness won't work

**END OF DISTRIBUTION IMPUTATION**

End of distribution imputation is a technique for handling missing data by replacing them with extreme values from the non-missing data distribution. Typically applied to numerical features, it calculates percentiles (e.g., 99th or 1st percentile) of available values, then substitutes missing data with these extreme values. This approach maintains data distribution integrity while addressing missing values. However, it may introduce bias if missingness is not random and can potentially inflate the significance of outliers. Careful consideration of data context and problem requirements is essential when deciding whether to use this imputation method.

```python
df = pd.read_csv('/content/tested.csv', usecols = ['Age','Fare','Survived'])
```

```python
df.head()
```

```python
df.Age.hist(bins = 50)
```

```python
extreme = df.Age.mean() + 3 * df.Age.std()
```

```python
# sns.boxplot('Age',data = df)
```

```python
def impute_NanEFDM(df, variable,median,extreme) :
  df[variable+"__end_distribution"] = df[variable].fillna(extreme)
  df[variable].fillna(median, inplace =True)
```

```python
impute_NanEFDM(df,'Age',df.Age.median(),extreme)
```

```python
df.head()
```

```python
df.Age.hist(bins = 50)
```

```python
df.Age__end_distribution.hist(bins = 50)
```

```python
plt.boxplot(df['Age'])
```

```python
plt.boxplot(df['Age__end_distribution'])
```

No outlier at the end

**ARBITRARY VALUE IMPUTATION**

It consists of replacing NaN by an Arbitrary value

```python
df = pd.read_csv('/content/tested.csv', usecols = ['Age','Fare','Survived'])
```

```python
df.head()
```

```python
def imputeArbitrary(df, variable) :
  df[variable+'_zero'] = df[variable].fillna(0)
  df[variable+'_hundred'] = df[variable].fillna(100)
```

```python
def impute_featureCategory(df, variable) :
  most_frq_category = df[variable].value_counts().index[0]
  df[variable].fillna(most_frq_category, inplace = True)
```
