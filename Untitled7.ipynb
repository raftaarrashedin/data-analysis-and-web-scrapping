```python
import time
def fib(n) :
  if n == 0 or n == 1 :
    return 1
  else :
    return fib(n-1) + fib(n-2)
```

```python
start = time.time()
fib(5)
print(time.time()- start)
```

```python
def fiblesstime(n,d) :
  if n in d :
    return d[n]
  else :
    d[n] = fib(n-1,d) + fib(n-2,d)
    return d[n]
```

```python
start = time.time()
fiblesstime(5,d)
print(time.time()- start)
```
