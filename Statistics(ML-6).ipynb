```python
import pandas as pd
import numpy as np

df= pd.DataFrame({'Name' : ['Sharique','Shagufta','Bakr','Zamima','Ahmad','Sadaf','Kishwar','Nahid'],
                  'Salary' : [435323,533334,534453,64453,563434,564443,322423,324254]})

print(df.head())
```

```python
#calculate the mean, median, mode, min, max
print('means : ', df['Salary'].mean())
print("median : ",df['Salary'].median())
print("median : ",df['Salary'].mode())

print("Min : ",df['Salary'].min())
print("Max : ",df['Salary'].max())

print("Range : ",df['Salary'].max() - df['Salary'].min())
```

```python
df['Salary'].describe()
```

```python
from pandas.errors import InvalidIndexError
# Histogram of salaries 

%matplotlib inline
import matplotlib.pyplot as plt

print(df)

salary = df['Salary']
salary.plot.hist(title = 'Salary Distribution', color = 'grey', bins = 25)
plt.axvline(salary.mean(), color = 'violet', linestyle = 'dashed', linewidth=2)
plt.axvline(salary.median(), color = 'red', linestyle = 'dashed', linewidth=2)
plt.show()
```
