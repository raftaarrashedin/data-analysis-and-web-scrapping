Q(2) --> Create a dictionary of your choice having 5 elements. Keys should be integer value


```python
# Create a dictionary of your choice having 5 elements. Keys should be integer value

class intDict(dict):
  def __init__(self) :
    self = dict()

  def add_5_Element(self, key, value) :
    self[key] = value

objectDict = intDict()

objectDict.add_5_Element(1, 'Raftaar')
objectDict.add_5_Element(2, 'Rashedin')
objectDict.add_5_Element(3, 'Aman')
objectDict.add_5_Element(4, 'Jedy')
objectDict.add_5_Element(5, 'Reman')
print(objectDict)
```
